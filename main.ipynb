Converting column into 0/1

```python
df['Attrition'].value_counts()
```

```python
df['Attrition'] = df['Attrition'].apply(lambda x:1 if x == "Yes" else 0)
```

```python
df['Gender'].value_counts()
```

```python
df['Gender'] = df['Gender'].apply(lambda x:1 if x == "Male" else 0)
```

```python
df['Over18'].value_counts()
```

```python
df['Over18'] = df['Over18'].apply(lambda x:1 if x == "Y" else 0)
```

```python
df['OverTime'].value_counts()
```

```python
df['OverTime'] = df['OverTime'].apply(lambda x:1 if x == "Yes" else 0)
```

```python
df.head()
```

One-Hot Encoding      
Converts categorical columns into dummy columns.

```python
cols = ["BusinessTravel","Department","EducationField","JobRole","MaritalStatus"]
```

```python
df = pd.get_dummies(df, columns=cols)
```

```python
df=df.drop(columns=["BusinessTravel","Department","EducationField","JobRole","MaritalStatus","EmployeeNumber"],axis=1,errors='ignore').astype(int)
```

```python
df
```

```python
df.hist(figsize=(20,15))
plt.tight_layout()
plt.show()
```

Drop unnecessary constant columns    
Not useful for ML, so dropped

```python
df = df.drop(["EmployeeCount","Over18","StandardHours"],axis=1)
```

```python
X = df.drop('Attrition',axis=1)
y = df["Attrition"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2,random_state=26)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_jobs=1)
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
model.score(X_test,y_test)
```

Creates a sorted dictionary of feature name, importance score

```python
sort_importance =dict(sorted(zip(model.feature_names_in_, model.feature_importances_),key=lambda x:x[1],reverse=True))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,8))
plt.bar(sort_importance.keys(), sort_importance.values())
plt.xticks(rotation = 45, ha = "right")
plt.show()
```
